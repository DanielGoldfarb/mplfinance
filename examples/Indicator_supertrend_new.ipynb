### Supertrend Formula

The Supertrend Indicator requires that we first calculate an UpperBand and a LowerBand.   

The Supertrend itself is then a single trend line that is, based on the rules described below, sometimes the upper band, sometimes the lower band, and sometimes a constand value.

The Upper and Lower bands are calculated as follows:

```python
UpperBand = MedianPrice + (Multiplier * ATR(Periods))
LowerBand = MedianPrice - (Multiplier * ATR(Periods))
```

where 
- **`MedianPrice`** at each period is `(High+Low)/2`, and

- **`ATR`** is the `Average True Range`, a rolling average of the `True Range` for some number of Periods,
  - The `True Range` at each period is the maximum of: High-Low, |High-PreviousClose|, |Low-PreviousClose|
  
- the **Multiplier** and number of **Periods** are adjustable parameters that the user can provide.

````python
pub = tdf.ub[0]
plb = tdf.lb[0]
pst = tdf.lb[0]
st  = 'lower'
super_trend[0] = tdf.lb[0]

for ix in range(1,len(tdf)):
    cp = tdf.Close[ix]
    if st == 'lower':
        if cp < pst:
            st = 'upper'
            super_trend[ix] = tdf.ub[ix]
            pub = tdf.ub[ix]
        elif tdf.lb[ix] < plb:
            super_trend[ix] = plb
        else:
            super_trend[ix] = tdf.lb[ix]
            plb = tdf.lb[ix]
    elif st == 'upper':
        if cp > pst:
            st = 'lower'
            super_trend[ix] = tdf.lb[ix]
            plb = tdf.lb[ix]
        elif tdf.ub[ix] > pub:
            super_trend[ix] = pub
        else:
            super_trend[ix] = tdf.ub[ix]
            pub = tdf.ub[ix]
    else:
        raise ValueError('st must be `lower` or `upper`')
    pst = super_trend[ix]
````

````python
def calculate_atr(atr_length, highs, lows, closes):
    """Calculate the average true range
    atr_length : time period to calculate over
    all_highs : list of highs
    all_lows : list of lows
    all_closes : list of closes
    """
    if atr_length < 1:
        raise ValueError("Specified atr_length may not be less than 1")
    elif atr_length >= len(closes):
        raise ValueError("Specified atr_length is larger than the length of the dataset: " + str(len(closes)))
    atr = 0
    for i in range(len(highs)-atr_length, len(highs)):
        high = highs[i]
        low = lows[i]
        close_prev = closes[i-1]
        tr = max(abs(high-low), abs(high-close_prev), abs(low-close_prev))
        atr += tr
        print('atr=',atr)
    return atr/atr_length
````

````python
#Function For Color Coding Supertrend
def color(super_trend):
    UP = []
    DOWN = []
    for i in range(len(super_trend)):
        if super_trend['SUPERTl_7_3.0'][i] < super_trend['Close'][i]:
            UP.append(int(super_trend['SUPERTl_7_3.0'][i]))
            DOWN.append(np.nan)
        elif super_trend['SUPERTl_7_3.0'][i] > super_trend['Close'][i]:
            DOWN.append(int(super_trend['SUPERTs_7_3.0'][i]))
            UP.append(np.nan)
        else:
            UP.append(np.nan)
            DOWN.append(np.nan)
    super_trend['up'] = UP
    super_trend['down'] = DOWN
    return super_trend        
````
