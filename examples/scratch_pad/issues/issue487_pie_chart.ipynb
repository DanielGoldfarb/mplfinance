```python

import yfinance as yf
import pandas_datareader.data as data
from datetime import datetime, timedelta
```

```python
end_date = datetime.today()
start_date = datetime.today() - timedelta(days=150)
df = data.get_data_yahoo('SPY', start_date, end_date)
```

```python
pie_data = [7,20,11,13]
fig, axlist = mpf.plot(df,type='candle', returnfig=True, mav=(20,50), tight_layout=False, 
                       figsize=(18,7), volume=False, title='SPY', 
                       scale_padding={'left': 1, 'top': -0.1, 'right': 0.85, 'bottom': 1})
```

```python
pie_data = [7,20,11,13]
pie_ax = fig.add_axes([0.1,0.6,0.35,0.35])

pie_ax.pie(pie_data, autopct='%.1f', textprops={'fontsize': 13}, 
           wedgeprops={"linewidth": 2, "edgecolor": "white"},
          radius=1.25)
fig
```
