```python

s3 = [[('2021-03-04', 373.260009765625) , ('2021-05-17', 373.260009765625)], 
      [('2021-03-25', 385.3399963378906), ('2021-05-17', 385.3399963378906)], 
      [('2021-05-12', 406.3399963378906), ('2021-05-17', 406.3399963378906)]
     ]

r3 = [[('2021-03-17', 399.6499938964844) , ('2021-05-17', 399.6499938964844)], 
      [('2021-04-16', 419.4800109863281) , ('2021-05-17', 419.4800109863281)], 
      [('2021-04-23', 419.8399963378906) , ('2021-05-17', 419.8399963378906)], 
      [('2021-05-07', 424.42999267578125), ('2021-05-17', 424.42999267578125)]
     ]


print()
print('[s3,r3]=',[s3,r3])
print()
print('s3+r3=',s3+r3)
print()


import pandas as pd

df = pd.read_csv('../data/yahoofinance-SPY-20200901-20210113.csv',index_col=0,parse_dates=True)
```

```python
df.head(6)
len(df)
```

```python
ix = pd.bdate_range('2021-03-01','2021-05-23',freq='B')
len(ix)
```

```python
tdf = df.iloc[8:68,:]
newdf = pd.DataFrame(dict(Open=tdf.Open.values,High=tdf.High.values,Low=tdf.Low.values,Close=tdf.Close.values),index=ix) + 50.0
```

```python
newdf.head(3)
newdf.tail(3)
```

```python
import mplfinance as mpf
```

```python
mpf.plot(newdf,type='candle',alines=dict(alines=s3,colors='g'))
mpf.plot(newdf,type='candle',alines=dict(alines=r3,colors='r'))
```

```python
g = ['g']*len(s3)
r = ['r']*len(r3)

mpf.plot(newdf,type='candle',alines=dict(alines=s3+r3,colors=g+r))
```
